```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
# Generate sample data
np.random.seed(0)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel()

# Add noise to targets
y[::5] += 3 * (0.5 - np.random.rand(20))
```

```python
# Train SVR model with rbf kernel
svr_poly = SVR(kernel='rbf', gamma='auto')
svr_poly.fit(X, y)
```

```python
# Generate points for the curve
X_curve = np.linspace(0, 5, 100)[:, np.newaxis]
y_pred = svr_poly.predict(X_curve)
```

```python
# Plot the data points
plt.scatter(X, y, color='black', label='Data')

# Plot the polynomial curve
plt.plot(X_curve, y_pred, color='red', lw=2, label='Polynomial Regression')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression using SVR Polynomial Kernel')
plt.legend()

# Show plot
plt.show()
```

```python
# Calculate R2 score
r2 = r2_score(y, svr_poly.predict(X))
print("r2 score : ",r2)
```
